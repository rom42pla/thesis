```python
import os
from os.path import join
import json
from pprint import pprint
import random

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from tqdm import tqdm
```

```python
PRINT_LATEX = True
```

# Pascal VOC dataset

## Images reading

```python
sizes_voc = []
for image_name in tqdm(os.listdir(join('datasets', 'VOC2012', 'JPEGImages')), desc=f"Reading images"):
    image = cv2.imread(join('datasets', 'VOC2012', 'JPEGImages', image_name))
    sizes_voc += [image.shape[:2]]

sizes_voc = np.asarray(sizes_voc)
```

## Pixels density

```python
print(f"Images in VOC2012: {len(sizes_voc)}")
print(f"Average number of pixels in an image {int(sizes_voc.prod(-1).mean())}")

bar, ax = plt.subplots(figsize=(16,8))
ax = sns.histplot(x=sizes_voc.prod(-1), palette="rocket", bins=20)
ax.set_title("Number of pixels distributions for PASCAL VOC2012")
ax.set_xlabel("pixels")
ax.set_ylabel("count")
plt.show()
```

## Sample images

```python
images = random.choices([i
                         for i in os.listdir(join("datasets", "VOC2012", "JPEGImages")) 
                         if i.endswith(".jpg")], 
                        k=9)
images = [cv2.imread(join("datasets", "VOC2012", "JPEGImages", i))[...,::-1]
          for i in images]

fig, axs = plt.subplots(3, 3, figsize=(2*9, 2*9))
for img, ax in zip(images, axs.flatten()):
    ax.imshow(img)
fig.suptitle("Sample images from PASCAL VOC2012")
plt.show()
```

```python
pascal_voc_labels = {
    label.split("_")[0] 
    for label in os.listdir(join("datasets", "VOC2012", "ImageSets", "Main"))
    if not label.split("_")[0].endswith(".txt")
}
print(f"Labels in VOC2012 ({len(pascal_voc_labels)}):")
pprint(pascal_voc_labels)
```

```python
labels_distribution = pd.DataFrame()

for filename in os.listdir(join("datasets", "VOC2012", "ImageSets", "Main")):
    if not ("_" in filename and filename.endswith("trainval.txt")):
        continue
    label, instances = filename.split("_")[0], 0
    with open(join("datasets", "VOC2012", "ImageSets", "Main", filename)) as fp:
        for line in fp:
            if line.strip().split(" ")[-1] == "1":
                instances += 1
    labels_distribution = labels_distribution.append({
        "label": label,
        "instances": instances
    }, ignore_index=True) 
labels_distribution_voc = labels_distribution.sort_values(by="instances").reset_index()
```

```python
bar,ax = plt.subplots(figsize=(16,8))
sample_indexes = np.linspace(0, len(labels_distribution_voc)-1, 10, dtype=int).tolist()
ax = sns.barplot(data=labels_distribution_voc.iloc[sample_indexes], x="label", y="instances", palette="rocket")
ax.set_title("PASCAL VOC2012 instances distributions (sample)")
ax.set_xlabel("label")
ax.set_ylabel("count")
plt.show()
```

# COCO

## Images reading

```python
sizes_coco = []
for t in ["train2017", "val2017"]:    
    with open(join('datasets', 'coco', "annotations", f"instances_{t}.json")) as fp:
        file = json.load(fp)
        sizes_coco += [(image_info["width"], image_info["height"]) 
                       for image_info in file["images"]]
    
sizes_coco = np.asarray(sizes_coco)
```

## Pixels density

```python
print(f"Images in COCO2017: {len(sizes_coco)}")
print(f"Average number of pixels in an image {int(sizes_coco.prod(-1).mean())}")

bar, ax = plt.subplots(figsize=(16,8))
ax = sns.histplot(x=sizes_coco.prod(-1), palette="rocket", bins=20)
ax.set_title("Number of pixels distributions for COCO2017")
ax.set_xlabel("pixels")
ax.set_ylabel("count")
plt.show()
```

## Sample images

```python
images = random.choices([i
                         for i in os.listdir(join("datasets", "coco", "train2017")) 
                         if i.endswith(".jpg")], 
                        k=9)
images = [cv2.imread(join("datasets", "coco", "train2017", i))[...,::-1]
          for i in images]

fig, axs = plt.subplots(3, 3, figsize=(2*9, 2*9))
for img, ax in zip(images, axs.flatten()):
    ax.imshow(img)
fig.suptitle("Sample images from COCO2017")
plt.show()
```

```python
for t in ["train2017", "val2017"]:
    print(f"Images in COCO ({t}): {len(os.listdir(join('datasets', 'coco', t)))}")
    
    with open(join('datasets', 'coco', "annotations", "instances_val2017.json")) as fp:
        coco_labels = {label_infos["id"]: label_infos["name"] 
                       for label_infos in json.load(fp)["categories"]}
    
    print(f"Labels in COCO ({t}) ({len(coco_labels)}):")
    pprint(coco_labels.values())
    print()
```

```python
if PRINT_LATEX:
    print(f"Latex friendly print:\n")
    template = "\
    \\begin{multicols}{3}\n\
        \\begin{itemize}\n\
    __items__\n \
        \\end{itemize}\n\
    \\end{multicols}\
    "
    print(template.replace("__items__", 
                           "".join(sorted(["\t\\item {}\n".format(label) for label in coco_labels])).rstrip()))
```

```python
labels_distribution_coco = {
    label: 0
    for label in coco_labels.values()
}

for filename in ["instances_train2017.json", "instances_val2017.json"]:
    with open(join('datasets', 'coco', "annotations", filename)) as fp:
        # reads the file
        file = json.load(fp)
        
    #retrieves the list of labels
    for annotation in file["annotations"]: 
        labels_distribution_coco[coco_labels[annotation["category_id"]]] += 1
        
        
        
# print(labels_distribution_coco)
labels_distribution_coco_df = pd.DataFrame()
for label, instances in labels_distribution_coco.items():
    labels_distribution_coco_df = labels_distribution_coco_df.append(
        {"label": label, 
        "instances": int(instances)},
        ignore_index=True)
labels_distribution_coco = labels_distribution_coco_df.sort_values(by="instances").reset_index()
```

```python
bar,ax = plt.subplots(figsize=(16,8))
sample_indexes = np.linspace(0, len(labels_distribution_coco)-1, 10, dtype=int).tolist()
ax = sns.barplot(data=labels_distribution_coco.iloc[sample_indexes], 
                 x="label", y="instances", palette="rocket")
ax.set_title("COCO2017 instances distributions (sample)")
ax.set_xlabel("label")
ax.set_ylabel("count")
plt.show()
```

# LVIS

```python
labels_distribution_lvis = {}

for t in ["train", "val"]:    
    # opens the file
    with open(join("datasets", "lvis", f"lvis_v0.5_{t}.json")) as fp:
        file = json.load(fp)
    # retrieves the list of labels
    lvis_labels = {label_infos["id"]: label_infos["name"] 
                   for label_infos in file["categories"]}
    # retrieves the distribution of labels
    labels_distribution_lvis = {
        label: 0
        for label in lvis_labels.values()
    }
    for annotation in file["annotations"]: 
        labels_distribution_lvis[lvis_labels[annotation["category_id"]]] += 1
    
# converts the labels to a dataframe
labels_distribution_lvis_df = pd.DataFrame()
for label, instances in labels_distribution_lvis.items():
    labels_distribution_lvis_df = labels_distribution_lvis_df.append(
        {"label": label, 
        "instances": int(instances)},
        ignore_index=True)
labels_distribution_lvis = labels_distribution_lvis_df.sort_values(by="instances").reset_index()
```

```python
labels_frequencies = {}
# retrieves rare, common and frequent labels
for t in ["rare", "common", "freq"]:
    # opens the file
    with open(join("datasets", "lvis", f"lvis_v0.5_train_{t}.json")) as fp:
        file = json.load(fp)
    labels_frequencies[t] = {lvis_labels[annotation["category_id"]]
                             for annotation in file["annotations"]}
pprint(labels_frequencies)
```
